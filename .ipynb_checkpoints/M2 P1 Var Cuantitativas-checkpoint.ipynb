```python
# Creación aleatoria de datos
import numpy as np
X = np.array([0.5, 23, 0.3, 4.5, 0.3, 0.5,
              -28, -50, 60, -100, -10, -11, 13, 19, 1, 9])
```

```python
# Medidas de tendencia central
from scipy import stats
media = np.mean(X)
mediana = np.median(X)
moda, _ = stats.mode(X)
print('media: ', media)
print('mediana: ', mediana)
print('moda :', moda)
```

```python
# Medidas de posición
import numpy as np
import matplotlib.pyplot as plt

Q1 = np.percentile(X, 25)
Q3 = np.percentile(X, 75)
RIC = Q3 - Q1
lim_inf = Q1 - 1.5*RIC
lim_sup = Q3 + 1.5*RIC

print('RIC :', RIC)
print('limite inferior :', lim_inf)
print('limite superior :', lim_sup)
```

```python
# Dibuja el diagrama de box and whiskers y determina visualmente el número de outliers
plt.boxplot(X)
plt.grid()
plt.title("Diagrama de caja y bigotes")
plt.show()
```

```python
# Medidas de dispersión
desvest = np.std(X)
print('Desviación típica: ', desvest)
print('Varianza: ', desvest**2)
```

```python
# Medidas de distribución
asimetria = stats.skew(X)
curtosis = stats.kurtosis(X, fisher=True) # https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.kurtosis.html
print('Asimetría ', np.round(asimetria, 4))
print('Curtosis ', np.round(curtosis, 4))

if curtosis>0:
    print('Distribución Leptocúrtica')
elif curtosis<0:
    print('Distribución Platicúrtica')
else:
    print('Distribución Mesocúrtica')
```

```python
# EJERCICIO EXTRA SOBRE EL DATASET DE IRIS
# Extraer la media y la desviación típica de cada variable en función de la clase y representarlo en un gráfico (plt.errorbar)

#1) Creamos un dataframe
#con feature_names hago que muestre el nombre de las columnas en iris
#Estos datos (iris.data) tienen estas columnas con estos nombres
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df["class"] = iris.target #los agrupo por clase

#2) Tengo que mapear la clase con sus nombres (asignar valores)

df["class"] = df["class"].map({
    0: "setosa",
    1: "versicolor",
    2: "virginica"
})

#3) Ahora voy a calcular la media y la desviación estandar por clase

media_iris = df.groupby("class").mean() #media
std_iris = df.groupby("class").std() #desviación estándar

#4) Toca graficar

plt.figure()

# Iterar por cada variable
for col in media_iris.columns:
    plt.errorbar(
        media_iris.index, # clases
        media_iris[col], # medias
        yerr=std_iris[col], # desviación estándar
        label=col,
        capsize=5,
        marker='o'
    )

plt.xlabel("Clase")
plt.ylabel("Valor")
plt.title("Ejercicio para resolver, iris")
plt.legend()
plt.grid()

plt.show()
```
